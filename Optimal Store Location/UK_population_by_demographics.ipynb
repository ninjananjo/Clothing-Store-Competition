```python
#https://data.gov.uk/dataset/b1f548a4-694c-438b-9551-af7a121e3395/ons-mid-year-population-estimates-custom-age-tables
link = 'https://data.london.gov.uk/download/ons-mid-year-population-estimates-custom-age-tables/11cff330-73d7-41ef-87c1-9b48337f98f4/ons-mye-custom-age-tool-uk-geogs-2015.xls'
population_estimates = pd.read_excel(link, sheet_name = 2, header=[1,2])
```

```python
population_estimates.columns = population_estimates.columns.map('{0[0]}-{0[1]}'.format) 
population_estimates = population_estimates.rename(columns={'Unnamed: 0_level_0-Code':'Code', 'Unnamed: 1_level_0-Borough':'Borough'})
population_estimates.drop(columns=['M-No Males','F-No Females'], inplace=True)
population_estimates.head()
```

```python
population_estimates_melt = pd.melt(population_estimates, id_vars=['Code','Borough'], value_vars=population_estimates.columns[2:], value_name='Persons')
population_estimates_melt['Gender'] = population_estimates_melt['variable'].str[0:1]
population_estimates_melt['Age'] = population_estimates_melt['variable'].str[2:]
population_estimates_melt.drop(columns=['variable'], inplace=True)
population_estimates_melt.set_index(['Code'], inplace=True)
population_estimates_melt.sample(10)
```
