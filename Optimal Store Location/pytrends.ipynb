```python
from pytrends.request import TrendReq
import pandas as pd
import requests
import lxml
import xlrd
```

```python
pytrend = TrendReq()

kw_list = ["Fashion"]
pytrend.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='GB', gprop='')
```

```python
interest_by_region_df = pytrend.interest_by_region(resolution='CITY', inc_low_vol=True)
print(interest_by_region_df.head())
```
