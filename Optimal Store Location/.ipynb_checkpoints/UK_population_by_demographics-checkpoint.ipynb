```python
#https://geoportal.statistics.gov.uk/datasets/local-authority-districts-december-2017-super-generalised-clipped-boundaries-in-united-kingdom-wgs84?geometry=-4.821%2C53.572%2C2.59%2C54.698
path = 'Shapefile/Local_Authority_Districts_December_2017_Super_Generalised_Clipped_Boundaries_in_United_Kingdom_WGS84.shp'
gdf = gpd.read_file(path)
```

```python
gdf.sample(5)
```

```python
gdf.plot(figsize=(16,8))
```

```python
#https://data.gov.uk/dataset/b1f548a4-694c-438b-9551-af7a121e3395/ons-mid-year-population-estimates-custom-age-tables
link = 'https://data.london.gov.uk/download/ons-mid-year-population-estimates-custom-age-tables/11cff330-73d7-41ef-87c1-9b48337f98f4/ons-mye-custom-age-tool-uk-geogs-2015.xls'
population_estimates = pd.read_excel(link, sheet_name = 2, header=[1,2])
```

```python
population_estimates.columns = population_estimates.columns.map('{0[0]}-{0[1]}'.format) 
population_estimates = population_estimates.rename(columns={'Unnamed: 0_level_0-Code':'Code', 'Unnamed: 1_level_0-Borough':'Borough'})
population_estimates.drop(columns=['M-No Males','F-No Females'], inplace=True)
population_estimates.head()
```

```python
population_estimates_melt = pd.melt(population_estimates, id_vars=['Code','Borough'], value_vars=population_estimates.columns[2:], value_name='Persons')
population_estimates_melt['Gender'] = population_estimates_melt['variable'].str[0:1]
population_estimates_melt['Age'] = population_estimates_melt['variable'].str[2:]
population_estimates_melt['Age'].replace('90+','90', inplace=True)
population_estimates_melt['Age'] = population_estimates_melt['Age'].astype('int32')
population_estimates_melt.drop(columns=['variable'], inplace=True)
population_estimates_melt.set_index(['Code'], inplace=True)
population_estimates_melt.sample(5)
population_estimates_melt.shape
```

```python
min_age =16
max_age =30
gender ='F'
#grouped_columns = ['Borough', 'objectid', 'lad17cd','bng_e', 'bng_n', 'long', 'lat', 'st_areasha', 'st_lengths','geometry']


population_gdf = population_estimates_melt.merge(gdf, left_on='Code', right_on='lad17cd')
population_gdf.drop(columns=['lad17nm','lad17nmw'], inplace=True)
population_gdf = population_gdf[population_gdf['Gender'] == gender]
population_gdf = population_gdf[(population_gdf['Age']>=min_age) & (population_gdf['Age']<=max_age)]
```

```python
import json
#Read data to json.
merged_json = json.loads(population_gdf.to_json())
#Convert to String like object.
json_data = json.dumps(merged_json)
```
